```python
#Changing data types
#purchase amount to float

df['Purchase_Amount'] = (
    df['Purchase_Amount'].str.replace('$', '', regex=False).astype(float)
)
```

```python
# Normalize separators
df['Time_of_Purchase'] = df['Time_of_Purchase'].astype(str).str.replace('/', '-', regex=False)

# Convert to datetime
df['Time_of_Purchase_dt'] = pd.to_datetime(
    df['Time_of_Purchase'],
    errors='coerce'
)

# Format for display (DD-MM-YYYY)
df['Time_of_Purchase'] = df['Time_of_Purchase_dt'].dt.strftime('%d-%m-%Y')

```

```python
#gender to category
df['Gender'] = df['Gender'].astype('category')

#income level to category
df['Income_Level'] = df['Income_Level'].astype('category')

#maritial status to category
df['Marital_Status'] = df['Marital_Status'].astype('category')

#educationlevel to category
df['Education_Level'] = df['Education_Level'].astype('category')

#occupation to category
df['Occupation'] = df['Occupation'].astype('category')

#Purchase_Category to category
df['Purchase_Category'] = df['Purchase_Category'].astype('category')

#Purchase_Channel as category
df['Purchase_Channel'] = df['Purchase_Channel'].astype('category')

#Social_Media_Influence as category
df['Social_Media_Influence'] = df['Social_Media_Influence'].astype('category')

#Discount_Sensitivity as category
df['Discount_Sensitivity'] = df['Discount_Sensitivity'].astype('category')

#Engagement_with_Ads as category
df['Engagement_with_Ads'] = df['Engagement_with_Ads'].astype('category')

#Device_Used_for_Shopping as category
df['Device_Used_for_Shopping'] = df['Device_Used_for_Shopping'].astype('category')

#Payment_Method as category
df['Payment_Method'] = df['Payment_Method'].astype('category')

#Purchase_Intent as category
df['Purchase_Intent'] = df['Purchase_Intent'].astype('category')

#Shipping_Preference as category
df['Shipping_Preference'] = df['Shipping_Preference'].astype('category')
```

```python
df.to_csv('cleaned_data.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Age vs Frequency of Purchase — Group By Analysis (bivariate)
#creating age group

bins = [18, 25, 35, 45, 55, 65]
labels = ['18–25', '26–35', '36–45', '46–55', '56+']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

age_group_analysis = df.groupby('Age_Group')['Frequency_of_Purchase'].agg(
    Count='count',
   
)

age_group_analysis

```

```python
#Income Level vs Purchase Amount

df.groupby('Income_Level')['Purchase_Amount'].agg(
    Average_Purchase='mean',
    Median_Purchase='median'
)

#Insight: Higher income levels are associated with higher purchase amounts.
```

```python
#Discount Used vs Purchase Frequency

df.groupby('Discount_Used')['Frequency_of_Purchase'].mean()

#Insight: Customers who use discounts tend to purchase more frequently.
```

```python
#1. Histogram of Purchase Amount

sns.histplot(df['Purchase_Amount'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Purchase Amount')
plt.xlabel('Purchase Amount ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
#3. Pie chart: Gender Distribution

gender_counts = df['Gender'].value_counts()

# Color palette
colors = sns.color_palette('pastel')

plt.figure(figsize=(5, 5))

plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    explode=[0.05] * len(gender_counts),   # slight separation
    shadow=True,
    wedgeprops={'edgecolor': 'white'},
    textprops={'fontsize': 9} 
)

plt.title('Gender Distribution', fontsize=14, fontweight='bold')
plt.axis('equal')  
plt.tight_layout()
plt.show()
```

###  Insights

```python
#4. Age Group vs Frequency of Purchase (barplot)


plt.figure(figsize=(9,5))
sns.barplot(
    data=df,
    x='Age_Group',
    y='Frequency_of_Purchase',
    hue='Gender',
    estimator=len,
    palette='Set2'
)
plt.title('Purchase Count by Age Group and Gender')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

```

```python
#5 Average Purchase Amount by Income Level & Marital Status (Heatmap)

# Set style
sns.set_theme(style="whitegrid")

plt.figure(figsize=(6,4))

sns.heatmap(
    pivot_table,
    annot=True,            # show values
    fmt=".2f",             # 2 decimal points
    cmap="YlGnBu",         # color map
    cbar_kws={'label': 'Average Purchase Amount'}
)

plt.title('Average Purchase Amount by Income Level & Marital Status', fontsize=14, fontweight='bold')
plt.ylabel('Income Level')
plt.xlabel('Marital Status')
plt.tight_layout()
plt.show()
```

### Insights
- Customers with **higher income levels** tend to have **higher average purchase amounts**.  
- **Married customers** generally spend more than single or divorced customers across most income levels.  
- The combination of **high income + married** appears to be the **highest spending segment**, making it a key target for premium offers or promotions.


```python
#6 Line chart: Total Purchase Amount Over Time

df['Time_of_Purchase'] = pd.to_datetime(df['Time_of_Purchase'], errors='coerce')
monthly_sales = df.groupby(df['Time_of_Purchase'].dt.to_period('M'))['Purchase_Amount'].sum()
monthly_sales.plot(kind='line', marker='o', color='purple')
plt.title('Monthly Total Purchase Amount Over Time')
plt.xlabel('Month')
plt.ylabel('Total Purchase Amount ($)')
plt.xticks(rotation=45)
plt.show()
```

###  Insights
- Purchase amount fluctuates monthly.

- Peaks may indicate seasonal trends (holidays, sales).

- Useful for marketing campaigns.

### Insights

- Middle income: More need-based and impulsive purchases

```python
#10 Payment Method vs Income Level

# Create crosstab
payment_income = pd.crosstab(df['Income_Level'], df['Payment_Method'])

# stacked bar chart
payment_income.plot(kind='bar', stacked=True)
plt.xlabel('Income Level')
plt.ylabel('Customer Count')
plt.title('Payment Method vs Income Level')
plt.xticks(rotation=0)
plt.legend(title='Payment Method')
plt.tight_layout()
plt.show()
```

### Insights

- Higher-income customers prefer digital and credit-based payment methods, while middle-income customers use more cash and debit payments.

##  **Insight Generation and Report**


## 🧩 Analysis Logic

The goal of this analysis was to explore customer behavior and identify meaningful patterns related to demographics, income levels, purchasing habits, decision-making behavior, and payment preferences.
Visual analytics was used to simplify complex relationships and support data-driven interpretations.

🔍 Key Insights and Interpretation


1️⃣ Customer Spending Behavior

- Most customers make purchases in the $200–$400 range, indicating moderate spending behavior.

- High-income customers, though fewer in number, tend to have higher average purchase amounts.





2️⃣ Demographic Patterns

- The customer base is dominated by middle-income individuals, who contribute the highest purchase volume.

- Female customers slightly outnumber male customers in the dataset.

- Customers aged 26–45 are the most active buyers, while younger (18–25) and older (56+) age groups show lower engagement.





3️⃣ Income Level and Purchase Behavior

- Middle-income customers exhibit higher purchase frequency, especially in essential categories such as food, groceries, and office supplies.

- High-income customers make fewer purchases but are more likely to spend on luxury and discretionary categories.




4️⃣ Purchase Intent and Decision-Making

- Middle-income customers show more need-based and impulsive purchasing behavior.

- High-income customers demonstrate more planned and wants-based purchases.

- A positive relationship exists between time spent on product research and time to decision, indicating thoughtful buying behavior among research-intensive customers.




5️⃣ Payment Preferences

- Higher-income customers prefer credit cards and digital payment methods.

- Middle-income customers rely more on cash and debit cards.

- This suggests increasing adoption of cashless payment systems with rising income levels.




6️⃣ Temporal and Seasonal Trends

- Purchase amounts fluctuate over time, with noticeable monthly peaks.

- These peaks may indicate seasonal trends, such as holidays or promotional periods, presenting opportunities for targeted marketing.

## ✅ Overall Summary and Conclusion


The analysis shows that customer purchasing behavior varies notably by income level. While product preferences remain largely consistent, differences emerge in purchase frequency, decision-making style, and payment methods. Middle-income customers contribute to higher transaction volume through frequent purchases, whereas high-income customers generate greater revenue per purchase through planned, high-value transactions.
